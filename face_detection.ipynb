```python
import cv2
print(cv2.__version__)
import matplotlib.pyplot as plt
```

```python
# from google.colab import drive
# drive.mount('/content/drive')
```

```python
cascade_classifier = cv2.CascadeClassifier('/content/haarcascade_frontalface_alt.xml')
```

```python
image = cv2.imread('/content/1b009f337c45c5bc380f7084582ca8f4.jpg')
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image)
```

```python
detected_faces = cascade_classifier.detectMultiScale(image, scaleFactor=1.1, minNeighbors=3, minSize=(20, 20))
print(detected_faces)
```

```python
for (X, Y, width, height) in detected_faces:
  cv2.rectangle(image, (X, Y), (X + width, Y + height), (0, 0, 255), 10)

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```
